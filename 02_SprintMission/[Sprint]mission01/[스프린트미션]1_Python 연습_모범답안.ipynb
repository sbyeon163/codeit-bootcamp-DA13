### 문제3. 모음 제거하기

- **문제 설명**  
문자열에서 모음(a, e, i, o, u)을 제거한 문자열을 반환하는 함수를 작성하세요.

- **함수 설명**  
`remove_vowels(s: string)`
  - s: 영어 소문자 알파벳으로 이루어진 문자열입니다.

- **입출력 예시**   

    - 입력1:

    ```
    "codeit"
    ```

    - 출력1:

    ```
    "cdt"
    ```

    - 입력2:
    
    ```
    "python programming"
    ```

    - 출력2:

    ```
    "pythn prgrmmng"
    ```



````python
def remove_vowels(s):
    vowels = "aeiou"
    result = ""

    for i in s:
        if i not in vowels:
            result += i

    return result
````

````python
# 프로그램 실행
remove_vowels("codeit")
````

````python
def remove_vowels(s):
    vowels = ['a', 'e', 'i', 'o', 'u']
    result = list(s)

    for v in vowels:
        if v in result:
            result.remove(v)
    result = ''.join(result)

    return result
````

````python
remove_vowels("codeit")
````

### 문제4. 수 크기 비교하기

- **문제 설명**  
두 개의 숫자를 입력받아, 큰 숫자, 작은 숫자 순서대로 숫자들을 반환하는 함수를 작성하세요.
단, 동일한 숫자를 2회 입력하는 경우에는 동일한 숫자 2개를 반환하면 됩니다.

- **함수 설명**

  `compare_numbers(num1, num2)`:  
  - num1, num2 : 정수형

- **입출력 예시**   

    - 입력1:

    ```
    20, 30
    ```

    - 출력1:

    ```
    30, 20
    ```

    - 입력2:
    
    ```
    10, 10
    ```

    - 출력2:

    ```
    10, 10
    ```

````python
def compare_numbers(num1, num2):
    if num2 > num1:
        return num2, num1
    else:
        return num1, num2
````

````python
# 프로그램 실행
print(compare_numbers(20, 30))
print(compare_numbers(10, 10))
````

````python
def compare_numbers_2(num1, num2):
    lst = [num1, num2]
    if num2 > num1:
        lst.sort(reverse=True)
    return tuple(lst)
````

````python
print(compare_numbers_2(20, 30))
print(compare_numbers_2(10, 10))
````

### 문제6. 총 합계 구하기

- **문제 설명**  
0부터 사용자가 입력한 숫자 n까지의 합을 구하는 함수를 작성하세요.

- **함수 설명**  
`sum_up_to_n(n)`:  
  - n: 0보다 큰 정수

- **입출력 예시**   

    - 입력1:

    ```
    3
    ```

    - 출력1:

    ```
    6
    ```

    - 입력2:
    
    ```
    10
    ```

    - 출력2:

    ```
    55
    ```



````python
def sum_up_to_n(n):
    i = 1
    total = 0

    while i <= n:
        total += i
        i += 1

    return total
````

````python
# 프로그램 실행
sum_up_to_n(3)
````

### 문제7. 짝수 합 구하기

- **문제 설명**

  숫자 n까지의 짝수 합을 구하는 함수를 작성하세요.

- **함수 설명**

  `sum_even_numbers()`

- **입출력 예시**

    - 입력:

    ```
    9
    ```

    - 출력:

    ```
    "1부터 9까지의 짝수 합은 20입니다."
    ```

````python
def sum_even_numbers(n):
    even_sum = sum(i for i in range(n+1) if i % 2 == 0)
    print(f"1부터 {n}까지의 짝수 합은 {even_sum}입니다.")
````

````python
# 프로그램 실행
sum_even_numbers(9)
````

### 문제8. 두 숫자 나누기

- **문제 설명**

  나눗셈을 하는 함수를 작성하세요. 예외가 발생하면 "잘못된 입력입니다."를 반환하도록 해 주세요.

- **함수 설명**
  `divide_numbers()`
  - 두개의 숫자를 차례대로 입력 받습니다.
  - 첫번째 숫자를 두번째 숫자로 나눕니다.
  - 결과값을 반환합니다. 단, 예외가 발생하면 "잘못된 입력입니다."를 반환합니다.
  - `try-except` 구문을 활용하여 함수를 작성해보세요. 이 문법에 대해서 구글링을 해도 좋습니다.

- **입출력 예시**

- 입력1:
    ```
    42, 7
    ```

- 출력1:
    ```
    6
    ```

- 입력2:
    ```
    135, 0
    ```

- 출력2:
    ```
    "잘못된 입력입니다."
    ```

````python
def divide_numbers(num1, num2):
    try:
        result = num1 / num2 # int
        return result
    except:
        return "잘못된 입력입니다."
````

````python
# 프로그램 실행
print(divide_numbers(135, 0))
````

### 문제9. BMI 계산하기

- **문제 설명**

  사용자가 입력한 키, 몸무게 값을 가지고 BMI를 계산해서 비만도를 반환하는 함수를 작성하세요.

- **함수 설명**

  `check_obesity()`
  - 사용자에게 키, 몸무게 값을 차례대로 입력 받습니다. (키:cm 기준, 몸무게:kg 기준)
  - 아래 공식을 참고하여 BMI 값을 계산합니다.
  - $$ BMI = \frac{\text{체중 (kg)}}{\text{신장 (m)}^2} $$
  - BMI 값을 기준으로 아래와 같이 비만도를 분류하여 반환합니다.
    - 18.5 미만: '저체중'
    - 18.5 이상 23 미만: '정상'
    - 23 이상 25 미만: '과체중'
    - 25 이상: '비만'


- **입출력 예시**

- 입력1:
    ```
    175, 60
    ```

- 출력1:
    ```
    정상
    ```

- 입력2:
    ```
    180, 88
    ```

- 출력2:
    ```
    비만
    ```

````python
def check_obesity(height, weight):
    height_m = height / 100

    bmi = weight / (height_m ** 2)

    if bmi < 18.5:
        return "저체중"
    elif bmi < 23:
        return "정상"
    elif bmi < 25:
        return "과체중"
    else:
        return "비만"
````

````python
# 프로그램 실행
print(check_obesity(178, 64))  # 정상
print(check_obesity(180, 88))  # 비만
````
