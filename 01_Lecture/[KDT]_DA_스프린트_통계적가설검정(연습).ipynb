# 오차와 신뢰구간

## 오차

- 통계학적 의미의 오차(error)
- 표본을 통해 얻은 평균 또는 비율이 참값(모집단의 실제 값)에서부터 어떤 확률과 정도로 벗어나 있는지에 대한 지표
- 표본을 통해 모집단을 파악하고 싶지만, 표본의 평균이 모집단의 평균과 완벽히 일치하는 경우는 거의 없음
- 때문에, 표본을 통해 모집단이 존재할 만한 범위를 추정
- 이때 오차라는 개념이 필요

- 오차는 수집된 데이터의 수와 해당 데이터의 불규칙성에 따라 달라짐
- 평균의 표준오차(Standard Error)

$$
SE = \frac{s}{\sqrt{n}}
$$

- $s$는 수집된 데이터의 표준편차 => 데이터의 불규칙성
        -표준편차가 크다는 것은 데이터의 불규칙성도 크다는 것.
- $n$은 표본 크기 => 데이터의 수

- 데이터의 수가 많을 수록 = `표본의 크기가 커질수록`
- 수집된 데이터의 불규칙성이 적을 수록 = `표준편차가 작을수록`
- 표본평균의 변동폭이 달라진다 = `표준오차가 작아진다`

- 실무에서는 반복적으로 표본을 뽑아 조사하기는 힘듭니다
- 표본을 한 번 뽑아 평균을 계산
- 그 표본평균은 분포 내 어딘가에 위치하게 될 것이라고 가정
- 표준오차가 작을수록 모집단의 참값에 더 가까워질 수 있다

## 신뢰구간

- 신뢰구간으로 참값의 범위 추정하기
- 표본 데이터를 통해서, 모집단의 실제값은 ~할 것이다 라고 주장
- 이 때, 신뢰구간이라는 개념이 필요

- 신뢰구간(CI, Confidence Interval)
: 모집단의 값이 어느 범위에 있는지 확률적으로 표현하는 방법
- 95% 신뢰구간이라는 표현을 자주 사용
- 95% 확률로 해당 구간에 모집단의 값이 포함되어 있다는 의미를 가짐
- 표본평균이 정규분포를 따를 때,
- 표본평균 +- 1.96SE 공식으로 계산
- 표본을 통해 얻은 표본평균값에서 1.96SE 범위 내에 모평균이 들어갈 확률이 95% 라는 의미

````python
import scipy.stats as stats
````

````python
np.random.seed(30)
sample_size = 30
pop_mean = 50
pop_std = 10

# 모집단에서 랜덤 표본 생성
sample = np.random.normal(pop_mean, pop_std, sample_size)
````

````python
sample  #30개의 표본 추출.
````

````python
# 표본 평균과 표본 표준편차
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1) #자유도 = n-1로 하나 깎인 상태로 둬야 함.
#ddof -> 자유도 , n-() 몇으로 나눠주는가?
````

````python
sample_mean
````

````python
sample_std
````

````python
# 신뢰 구간 계산 (95% 신뢰수준)
conf_level = 0.95
alpha = 1 - conf_level #0.05
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1) #df: 자유도 | 양측검정이므로 alpha를 2로 나눠줌. 

# 내 가설이 맞을까?
error = t_critical * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - error, sample_mean + error)
````

````python
error
````

````python
# 결과
print(f'표본 평균: {sample_mean:.2f}')
print(f'신뢰 구간: {confidence_interval[0]: .2f} ~ {confidence_interval[1]:.2f}')
````

## [2] 대응 표본 t-검정(Paired Sample t-test)
대응 표본: 전후 관계가 있는 표본

- 가정 : 모집단이 정규분포를 따른다.
- 검정 : 단일 모집단에 대해 처리를 가했을 때, 두 조건(전 vs 후, 처치 A vs B 등)의 평균 차이를 검정한다.
  - 귀무가설 : 두 모평균 차이가 0이다.
  - 대립가설 : 두 모평균 차이가 0이 아니다.

- 두 그룹이 각각 비정규성을 띄더라도 차이값의 샘플 분포가 정규성을 띄면 사용 가능
- 전, 후 그룹의 차이의 분포를 확인하는게 목적

### 예 : 정규성을 만족하는 경우

예) 10명의 참가자에게서 학습 효과를 확인하기 위해 학습 프로그램 참여 전과 후의 점수를 비교하였다. 각 참가자의 학습 프로그램 참여 전과 후 평균 점수에 차이가 있는지 검정하세요.(양측 검정, 유의수준=0.05)

````python
import pandas as pd
# 데이터 불러오기
exam_dict = {'before' : [60, 62, 65, 63, 67, 70, 72, 68, 66, 64],
        'after' : [63, 65, 69, 66, 71, 73, 77, 72, 68, 69]}

exam_df = pd.DataFrame(exam_dict)
exam_df
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다. | 모집단이 정규분포인 데이터로부터 샘플링됐다. 
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
exam_df['diff'] = exam_df['after'] - exam_df['before']
````

````python
from scipy import stats
s, p = stats.shapiro(exam_df['diff']) # 차이의 평균이 0인지 아닌지를 검저아는 것. 
print("diff:", p) # 0.24 : 유의수준 0.05보다 큼, 귀무가설 기각 못함, 정규성 따름
s, p = stats.shapiro(exam_df['after'])
print("after:", p)     # print("", 변수) --> , 꼭 써주기
s, p = stats.shapiro(exam_df['before'])
print("before:", p)
````

#### 대응 표본 검정

- 귀무가설 : D = 0 | D: 전후의 차이
- 대립가설 : D != 0

````python
from scipy import stats
stats.ttest_rel #ttest에서 두 개가 related 된 것을 검정. 
#x, y가 같은 값을 가져야 대응됨. 
# alternative : two-sided(Default) / less, greater: 단측검정 -> 대립가설을 기준
````

````python
stats.ttest_rel(exam_df['before'], exam_df['after']) #양측검정
print(p) #귀무가설 기각 / 대립가설 채택 -> 차이가 있다. 
````

단측검정
- 귀무가설 : D <= 0
- 대립가설 : D > 0

````python
s, p = stats.ttest_rel(exam_df['before'], exam_df['after'], alternative= 'greater') #단측검정 / 대립가설을 기준으로 설정
print(p) # 귀무가설 기각 못함 / before가 더 작다가 유의미
````

### 예 : 정규성을 만족하지 않는 경우

예) 10명의 참가자가 두 가지 수면 보조제(A, B)를 복용했을 때의 수면 시간 증가량(단위=시간)이 기록되어 있다. 참가자별로 두 가지 수면 보조제(A, B)를 복용했을 때의 수면 시간 증가량을 비교하여, 보조제 간에 차이가 있는지 통계적으로 검정하세요.(양측 검정, 유의수준=0.05)

````python
# 데이터 불러오기
sleep_dict = {'A' : [0.7, -1.6, -0.2, -1.2, -0.1, 3.4, 3.7, 0.8, 0, 2],
        'B' : [1.9, 0.8, 1.1, 0.1, -0.1, 4.4, 5.5, 1.6, 4.6, 3.4]}

sleep_df = pd.DataFrame(sleep_dict)
sleep_df['diff'] = sleep_df['A'] - sleep_df['B'] #차이를 나타내는 컬럼을 데이터프레임 내 추가
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
s,p = stats.shapiro(sleep_df['diff'])
print(p) # 귀무가설 기각 : 데이터가 정규분포를 따르지 않는다. 
````

#### 윌콕슨 부호 순위 검정

- 귀무가설 : D = 0 (차이가 없음)
- 대립가설 : D != 0 (차이가 있음)

````python
# wilcoxon(비모수 검정)--> 1) 정규성을 가정하지 않음(분포와 무관) 2) 차이값들의 순위와 중앙값 3) 이상치의 영향을 적게 받아 더 견고함. 
s, p = stats.wilcoxon(sleep_df['A'], sleep_df['B'])
print(p) # 귀무가설 기각 ; 차이가 있다. 
````

````python
# t-test(모수 검정) --> -> 1) 모집단이 정규분포를 따름(정규성 가정이 필수) 2) 두 집단 간의 평균 차이 3) 이상치에 민감 
s, p = stats.ttest_rel(sleep_df['A'], sleep_df['B'])
print(p) # 귀무가설 기각 ; 차이가 있다. 
````

- p-value < 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 있다(두 가지 수면 보조제(A, B) 간에 차이가 있다.)

## [3] 독립 표본 검정(Independent Samples t-test)

- 가정 : 정규성, 등분산성
- 검정 : 두 개의 독립된 모집단 평균을 비교
- 특징 : 두 그룹의 원소수가 달라도 됨 / 가정이 2가지

### 예 : 정규성과 등분산성을 만족하는 경우

예) 한 연구자는 두 가지 학습 방법(A, B)이 학생들의 시험 점수에 차이를 주는지 알아보고자 한다. 학습 방법(A, B)에 따른 평균 시험 점수에 차이가 있는지 검정하세요.(양측 검정, 유의수준=0.05)

- 귀무가설 : 학습 방법(A, B)에 따른 평균 시험 점수에 차이가 없다.
- 대립가설 : 학습 방법(A, B)에 따른 평균 시험 점수에 차이가 있다.

````python
# 데이터 준비하기
import pandas as pd
import numpy as np

exam_dict = {'group_a' : [54, 52, 46, 61, 42, 43, 54, 52, 46, 51, 45, 53, 45, 55, 53, 55, 57, 47, 49, 56, 49, 52, 50, 56, 49, 47, 60, 54, 55, 55],
           'group_b' : [48, 51, 51, 56, 58, 51, 57, 54, 49, 52, 51, 49, 64, 65, 58, 60, 55, 61, 49, 51, 63, 47, 56, 45, 57, 46, 59, 62, 60, 56]}

exam_df = pd.DataFrame(exam_dict)
exam_df.head()
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
# 독립 표본의 경우 각 그룹의 정규성을 확인해야 한다
from scipy import stats
s1, p1 = stats.shapiro(exam_df['group_a'])
s2, p2 = stats.shapiro(exam_df['group_b'])

print(p1,p2) #각 집단은 정규성을 따른다. 
````

#### 등분산성 검정

- 귀무가설 : 등분산을 따른다.
- 대립가설 : 등분산을 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 등분산을 따른다고 할만한 충분한 근거가 있다)

````python
# levene(레빈 검정): 여러 모집단 간의 분산이 동일한지(등분산성)을 확인하는 도구임.   
s, p = stats.levene(exam_df['group_a'], exam_df['group_b'])
print(p) #등분산을 따른다. 
````

#### 독립 표본 검정

````python
# ttest_ind : 독립 표본 검정
stats.ttest_ind(exam_df['group_a'], exam_df['group_b'], equal_var = True) # equal_var 등분산
print(p)
# equal_var = True : 스튜던트 t 검정
# equal_var = False : (정규성은 만족하나, 등분산을 만족하지 못함) : 웰치 t 검정 
````

````python
stats.ttest_ind(exam_df['group_a'], exam_df['group_b'], equal_var = False) 
print(p)
````

## 예 : 정규성과 등분산성을 만족하는 경우

예) iris 데이터에서 세 가지 붓꽃 품종별(setosa, versicolor, virginica)로 평균 sepal_width에 차이가 있는지 검정하세요.(유의수준=0.05)

- 귀무가설 : 세 가지 종의 sepal_width의 평균은 모두 같다.
- 대립가설 : 세 가지 종의 sepal_width의 평균이 모두 같은 것은 아니다.

````python
import seaborn as sns
iris_df = sns.load_dataset('iris')
display(iris_df.head(2))

print(iris_df['species'].unique())

# 세 가지 품종

setosa = iris_df[iris_df['species'] == 'setosa']['sepal_width']
versicolor = iris_df[iris_df['species'] == 'versicolor']['sepal_width']
virginica = iris_df[iris_df['species'] == 'virginica']['sepal_width']

# 정규성 검정 -> shapiro
from scipy import stats
print(stats.shapiro(setosa)[1]) # 튜플의 2번째 값(p-value)를 출력해라
print(stats.shapiro(versicolor)[1])
print(stats.shapiro(virginica)[1])

# 등분산 검정 -> levene
print(stats.levene(setosa, versicolor, virginica)[1]) # 3개 이상의 그룹도 하나하나 넣으면 됨 -> 등분산 만족
````

scipy.stats.shapiro 함수는 항상 (검정 통계량, p-value) 순서로 두 개의 값을 반환.
- **s(statistic)** ; shapiro(변수)[0] : **데이터가 얼마나 정규분포에 가까운지**를 나타내는 수치. **1에 가까울수록 완벽한 정규분포**를 의미.
- **p(p-value)** ; shapiro(변수)[1] : **귀무가설($H_0$)이 참일 확률.** 우리가 의사결정을 내릴 떄 가장 중요하게 보는 값.

샤피로 검정은 다른 가설 검정들과 귀무가설의 방향이 조금 다름

- 귀무가설 ($H_0$): 데이터가 정규분포를 따른다.
- 대립가설 ($H_1$): 데이터가 정규분포를 따르지 않는다.

- 판정 기준:
    - $p > 0.05$ 이면: 귀무가설을 채택하여 **정규성을 만족한다**고 결론 내림. (모수 검정 가능)
    - $p < 0.05$ 이면: 귀무가설을 기각하여 **정규성을 만족하지 않는다**고 결론 내립니다. (비모수 검정 고려)

### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

### 등분산성 검정

- 귀무가설 : 등분산을 따른다.
- 대립가설 : 등분산을 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 등분산을 따른다고 할만한 충분한 근거가 있다)

````python
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(data=program_df['value'], groups = program_df['program'])
tukeyhsd = mc.tukeyhsd(alpha = 0.05)
tukeyhsd.summary()
# a-c만 그룹간의 차이가 있다고 확인 / a-b 차이없음 / b-c 차이없음
````

````python
import pandas as pd
housing_df = pd.read_csv('kc_house_data.csv')
housing_df.head()
````

## 단순 선형 회귀

````python
x = housing_df[['sqft_living']] #데이터프레임에 대괄호 2개로 컬럼을 선택하면 데이터프레임이 나온다. 
y = housing_df[['price']]
````

````python
import matplotlib.pyplot as plt
plt.scatter(x,y,s=1)
plt.show()
````

````python
# 모델링 
# 방법 1 -  R언어 방식의 회귀식 구하기, ols : 최소제곱방법식으로 구한 회귀
from statsmodels.formula.api import ols
lr = ols('price ~ sqft_living', data = housing_df).fit() # ~ 앞에는 종속변수 / ~ 뒤에는 독립변수가 들어감|.fit() -> 모델 적용
#lr.summary() # 한번에 모델 결과 출력
#상수를 따로 관리 안 해도 된다는 장점이 있다. -> 독립변수가 100면 하나하나 다 입력해줘야 하니 복잡해지기 때문.
````

````python
# 회귀식 자체에 대한 통계값
lr.summary().tables[0]

# R-squared : 결정계수(0~1) / 종속변수의 변동성을 표현. 1에 가까울수록 선형성을 가진다. 0에 가까울 수록 설명력이 낮은.
# 0.493 : 49.3%의 변동성을 설명한다. 
# F-statistic : 회귀식의 f 통계량
# Prob (R-statistic) : pvalue -> 귀무가설 : 회귀식이 선형성을 가지지 않는다. / 귀무가설을 기각, 회귀식이 유의하다. 
````

````python
# 변수의 회귀식에 대한 유효성 - 각 독립변수들과 회귀계수, pv값을 확인 가능
lr.summary().tables[1]

# intercept : 절편
# coef : 계수들 or 절편값
# t : 검정통계량
# P>|t| : 회귀계수의 pvalue --> 0.05 이하면 귀무가설(회귀계수는 0이다)를 기각, 대립가설 채택 --> 유효한 회귀계수다.
# 회귀식을 정리하면
# Price = 280.6236 * sqft_living - 43580
````

````python
# 회귀식의 잔차의 유효성 --> 수업에서 다루지는 않음 / 깊이 있는 회귀에 대한 분석을 원하면 찾아보기
# lr.summary().tables[2]
````

````python
# ols 회귀방법 2 -> (필수) 특징은 상수항을 추가해서 값을 계산해준다는 점이다. 
import statsmodels.api as sm
X = housing_df['sqft_living']  #독립변수를 데이터단에서 컨트롤.
display(X.head(1))
X = sm.add_constant(X) # (필수) 상수항 추가 / 상수 초기화 -> 최적화해서 절편 intercept가 된다.
y = housing_df['price']
lr2 = sm.OLS(y,X).fit()
display(lr2.summary().tables[1]) #formula.api(방법1)과 결과 동일
````

````python
# 만약 새로운 데이터가 들어온다면? 방 면적이 25000인 경우의 집값을 예측하면?
# lr -> 방법1
# lr2 -> 방법2

#lr
new_data = pd.DataFrame({'sqft_living': [2500]})
lr.predict(new_data)

# lr2
lr2.predict([1,2500]) # 1은 constant
````
