## 불린 인덱싱(마스킹)

- 정의: 조건식의 결과로 나온 `True/False`값을 이용해, 원하는 데이터만 선택하는 방법
- 원리: 조건을 만족하는 행(또는 열)은 `True`로 표시하고, 조건을 만족하지 않으면 `False`로 숨김
- 예시: 특정 값 이상인 데이터, 특정 조건을 만족하는 행만 필터링

(예시1)

```python
gdp_array = np.array([6610, 7637, 8127, 8885, 10385, 12565, 14403, 12398, 8282, 10672])
gdp_array
```

```python
#gdp 1000보다 큰 경우만 출력

gdp_array > 10000 #mask ;마스킹
```

```python
gdp_array[gdp_array > 10000]
```

(예시2)

```python
#두 가지 조건
#10000보다 크거나 8000보다 작은 경우
mask1 = gdp_array > 10000
mask2 = gdp_array < 8000

gdp_array[mask1 | mask2] # or == | , and == & -->numpy에서는 기호로 표현
gdp_array[(gdp_array > 1000) | (gdp_array < 8000)] #괄호로 각 마스크를 닫아줘야 함. 
```

(예시3)

```python
# 8000이상이고 10000 이하인 경우
mask1 = gdp_array >= 8000
mask1 = gdp_array <= 10000

gdp_array[mask1 & mask2]
```

**Axis 0 (행 방향/세로)**

- 데이터를 추가하면 위에서 아래로(행이 늘어남) 쌓이죠?

- 결과적으로 **'세로 합계'**가 나옵니다.

**Axis 1 (열 방향/가로)**

- 데이터를 추가하면 왼쪽에서 오른쪽으로(열이 늘어남) 옆으로 붙죠?

- 그래서 연산도 왼쪽에서 오른쪽으로 훑으며 지나갑니다.

- 결과적으로 **'가로 합계'**가 나옵니다

## (4) 데이터 일부 선택하기

**데이터 선택 방법**

1. `iloc` (integer location): 행/열 **번호**로 접근하여 정수 기반 인덱스를 사용해 데이터 선택
2. `loc` (location): 행/열 **이름**으로 접근하여 레이블(이름) 기반 인덱스를 사용해 데이터 선택
3. 대괄호(`[]`) 사용
    - 열 선택: `df['column1']` → `'column1'` 열 반환  
    - 여러 열 선택: `df[['column1', 'column2']]` → 두 개 이상의 열 반환  
    - 행 슬라이싱: `df[0:3]` → 0~2번째 행 반환 (정수 위치 기준)

```python
import pandas as pd
burger_df = pd.read_csv("burger.csv")
```

(방법1 : iloc 사용)

```python
burger_df.iloc[2,1] #df.iloc[행, 열]
```

```python
burger_df.iloc[[0,2],[0,1,3]] #[[행],[열]]
```

```python
burger_df.iloc[0:2,1:3]
```

## (7) 그룹화

- `groupby`: 데이터 프레임에서 특정 컬럼을 기준으로 그룹을 나눈 뒤, 각 그룹별로 집계(합계, 평균 등), 변환, 필터링 등을 수행

```python
#groupby
#10골 이상 선수 legend 컬럼에 yes/ 미만인 선수들은 no
df.loc[df['assists'] >= 10,'legends'] = 'yes'
df.loc[df['assists'] < 10,'legends'] = 'no'
df

# legends yes/ no 그룹의 goals 평균
# dataframe.groupby([그룹할 컬럼])[타겟컬럼].연산메소드

df.groupby(['legends'])['goals'].mean()
```
