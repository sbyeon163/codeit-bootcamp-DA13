### 문제3. 모음 제거하기

- **문제 설명**  
문자열에서 모음(a, e, i, o, u)을 제거한 문자열을 반환하는 함수를 작성하세요.

- **함수 설명**  
`remove_vowels(s: string)`
  - s: 영어 소문자 알파벳으로 이루어진 문자열입니다.

- **입출력 예시**   

    - 입력1:

    ```
    "codeit"
    ```

    - 출력1:

    ```
    "cdt"
    ```

    - 입력2:
    
    ```
    "python programming"
    ```

    - 출력2:

    ```
    "pythn prgrmmng"
    ```



````python
def remove_vowels(s):
    s_list = list(s)
    vowels  = ["a","e","i","o","u"] #대괄호 두 번 하면 덩어리채로 담김
    for i in s_list[:]: #리스트에서 하나씩 불러서 할 땐 반복문 짱 /처음부터 끝까지
        for j in vowels:
            if i == j:
                s_list.remove(i) #j와 같은 i 제거 
    return "".join(s_list) #어디까지 돌고 반환해야 하는가를 기준으로 들여쓰기 / #리스트를 다시 병합
````

````python
#빈 문자열로 넣고 for문으로 하나씩 돌림
#for문 다 돌리고 비어있는 것들을 뱉음. 
#for문으로 돌리면 원소들을 하나씩 출력한다.
#문자열끼리 비교 가능
def remove_vowels(s):
    empty_string =""
    vowels ="aeiou"
    for char in s: 
        if char not in vowels:
            empty_string += char
    return empty_string
````

````python
# 프로그램 실행
remove_vowels("codeit")
````

````python
remove_vowels("python programming")
````

- **문제 설명**  
두 개의 숫자를 입력받아, 큰 숫자, 작은 숫자 순서대로 숫자들을 반환하는 함수를 작성하세요.
단, 동일한 숫자를 2회 입력하는 경우에는 동일한 숫자 2개를 반환하면 됩니다.

- **함수 설명**

  `compare_numbers(num1, num2)`:  
  - num1, num2 : 정수형

- **입출력 예시**   

    - 입력1:

    ```
    20, 30
    ```

    - 출력1:

    ```
    30, 20
    ```

    - 입력2:
    
    ```
    10, 10
    ```

    - 출력2:

    ```
    10, 10
    ```

````python

def compare_numbers(num1, num2):
    num = [num1, num2]
    
    if num1 < num2: 
        return sorted(num,reverse=True) #정렬된 결과값을 주는 sorted 사용/리스트에서는 reverse=True  
    else:
        return num #print는 단순 화면 출력일 뿐, 함수 결과값으로 데이터 넘겨주려면 return키워드 써야 함. 
````

````python
#리턴에 변수 2개 가능
def compare_numbers(num1, num2):
    if num1 < num2:
        return num1, num2 #
    else: 
        return num2, num1 
````

````python
# 프로그램 실행
print(compare_numbers(20, 30))
print(compare_numbers(10, 10))
````

### 문제7. 짝수 합 구하기

- **문제 설명**

  숫자 n까지의 짝수 합을 구하는 함수를 작성하세요.

- **함수 설명**

  `sum_even_numbers()`

- **입출력 예시**

    - 입력:

    ```
    9
    ```

    - 출력:

    ```
    "1부터 9까지의 짝수 합은 20입니다."
    ```

````python
def sum_even_numbers(n):
    j = 0
    for i in range(1,n+1):
        if i % 2 == 0:
            j = j + i
    print(f"1부터 {n}까지 짝수 합은 {j}입니다.")
````

````python
# 프로그램 실행
sum_even_numbers(9)
````

````python
#효율적
def sum_even_numbers(n):
    even_sum = sum(i for i in range(n+1) if i % 2 == 0)
    print(f"1부터 {n}까지의 짝수 합은 {even_sum}입니다.")
````

### 문제9. BMI 계산하기

- **문제 설명**

  사용자가 입력한 키, 몸무게 값을 가지고 BMI를 계산해서 비만도를 반환하는 함수를 작성하세요.

- **함수 설명**

  `check_obesity()`
  - 사용자에게 키, 몸무게 값을 차례대로 입력 받습니다. (키:cm 기준, 몸무게:kg 기준)
  - 아래 공식을 참고하여 BMI 값을 계산합니다.
  - $$ BMI = \frac{\text{체중 (kg)}}{\text{신장 (m)}^2} $$
  - BMI 값을 기준으로 아래와 같이 비만도를 분류하여 반환합니다.
    - 18.5 미만: '저체중'
    - 18.5 이상 23 미만: '정상'
    - 23 이상 25 미만: '과체중'
    - 25 이상: '비만'


- **입출력 예시**

- 입력1:
    ```
    175, 60
    ```

- 출력1:
    ```
    정상
    ```

- 입력2:
    ```
    180, 88
    ```

- 출력2:
    ```
    비만
    ```

````python
def check_obesity(height, weight):
    BMI = weight / ((height/100)**2) #1m == 100cm #입력받은 cm를 m로 바꿔줘야 하므로 /100
    
    if BMI < 18.5: #가장 큰 것 먼저 해주면 필터링 간편
        return '저체중'
    elif BMI >= 18.5 and BMI < 23:
        return '정상'
    elif BMI >= 23 and BMI < 25: #등호 넣기...
        return '과체중'
    elif BMI > 25:
        return '비만'
````
