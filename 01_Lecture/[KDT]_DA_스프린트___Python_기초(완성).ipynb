## [2] 자료형

**자료형(Data Type)**
- 숫자(Number) : 수학적 계산이 가능한 모든 숫자형 데이터
  - 정수(Integer) : 소수점을 포함하지 않는 숫자형 예) 10, -5, 0
  - 실수(Floating Point) : 소수점을 포함하는 숫자형 예) 3.14, -0.001, 2.0
- 문자열(String) : 문자들의 나열로 텍스트 데이터 예) "Python!", "123"
- 불린(Boolean) : 참(True) 또는 거짓(False) 데이터 예) True, False
- 리스트(List) : 여러 값을 순서대로 저장할 수 있는 가변형(mutable) 컬렉션 예) [1, 2, 3], ["apple", "banana", "cherry"]
- 딕셔너리(Dictionary) : 키-값 쌍으로 데이터를 저장하는 컬렉션 예) {"name": "Alice", "age": 25}
- 튜플(Tuple) : 리스트와 비슷하지만 변경할 수 없는(immutable) 데이터 컬렉션 예) (1, 2, 3), ("apple", "banana", "cherry")

````python
# 숫자
a = 5
b = 3.14
print(a + b)
````

````python
# 정수
x = 10
y = -4
print(x * y)
````

````python
# 실수
pi = 3.14159
radius = 2.5  # 반경
area = pi * radius ** 2  # 원의 면적
print(area)
````

````python
# 문자열
name = "Claire"
greeting = "Hello, " + name
print(greeting)
````

````python
print("2" + "5")
````

````python
# 불린(Boolean)
print(7 > 3)
print(3 > 7)
````

````python
# 리스트
fruits = ["apple", "banana", "cherry"]
print(fruits)
fruits.append("orange")
print(fruits)
print(fruits[1])
````

````python
# 리스트(값 변경 시도)
fruits[1] = 'pineapple'
print(fruits)
````

````python
# 딕셔너리
person = {"name" : "Claire", "age" : 20}
print(person)
print(person["name"])
print(person["age"])
````

````python
# 튜플
point = (3, 4)
print(point)
````

````python
# 튜플(값 변경 시도)
point[0] = 1
print(point)
````

**숫자형**

- 숫자(Number) : 수학적 계산이 가능한 모든 숫자형 데이터
  - 정수(Integer) : 소수점을 포함하지 않는 숫자형
  - 실수(Floating Point) : 소수점을 포함하는 숫자형

1. 정수형과 정수형 간의 연산은 정수형
2. 실수형과 실수형 간의 연산은 실수형
3. 정수형과 실수형 간의 연산은 실수형
4. (예외) 나눗셈은 항상 실수형

- 정수형과 정수형 간의 연산은 정수형

````python
# 덧셈
print(4 + 7)

# 뺄셈
print(2 - 4)

# 곱셈
print(5 * 3)

# 몫
print(7 // 3)

# 나머지
print(7 % 3)

# 거듭제곱
print(2 ** 3)
````

- 실수형과 실수형 간의 연산은 실수형

````python
# 덧셈
print(4.0 + 7.0)

# 뺄셈
print(2.0 - 4.0)

# 곱셈
print(5.0 * 3.0)

# 몫
print(7.0 // 3.0)

# 나머지
print(7.0 % 3.0)

# 거듭제곱
print(2.0 ** 3.0)
````

- 정수형과 실수형 간의 연산은 실수형

````python
# 덧셈
print(4 + 7.0)

# 뺄셈
print(2.0 - 4)

# 곱셈
print(5.0 * 3)

# 몫
print(7 // 3.0)

# 나머지
print(7 % 3.0)

# 거듭제곱
print(2.0 ** 3)
````

- (예외) 나눗셈은 항상 실수형

````python
print(7 / 2)
print(7.0 / 2)
print(6.0 / 2.0)
````

#### 실습) 문자열 포매팅 실습

[링크](https://www.codeit.kr/topics/core-concept-of-python-programming/lessons/2757)

(f-string 사용)

````python
wage = 5  # 시급 (1시간에 5달러)
exchange_rate = 1142.16  # 환율 (1달러에 1142.16원)

# f-string 사용
print(f"{1}시간에 {wage * 1}달러 벌었습니다.")
print(f"{5}시간에 {wage * 5}달러 벌었습니다.")
print(f"{1}시간에 {wage * 1 * exchange_rate:.1f}원 벌었습니다.")
print(f"{5}시간에 {wage * 5 * exchange_rate:.1f}원 벌었습니다.")
````

(format 메소드 사용)

````python
wage = 5  # 시급 (1시간에 5달러)
exchange_rate = 1142.16  # 환율 (1달러에 1142.16원)

# format 메소드 사용
print("{}시간에 {}{} 벌었습니다.".format(1, wage * 1, "달러"))
print("{}시간에 {}{} 벌었습니다.".format(5, wage * 5, "달러"))
print("{}시간에 {:.1f}{} 벌었습니다.".format(1, wage * 1 * exchange_rate, "원"))
print("{}시간에 {:.1f}{} 벌었습니다.".format(5, wage * 5 * exchange_rate, "원"))
````

#### 형 변환
- 형 변환(Type Conversion, Type Cast) : 한 자료형에서 다른 자료형으로 바꾸는 것
- 한 자료형에서 다른 자료형으로 바꾸는 형 변환을 하는 경우 논리적으로 타당한지 고려가 필요하다.

(예시)
- 숫자형 -> 숫자형 변환
  - `int()` : 정수형으로 변환
  - `float()` : 실수형으로 변환

````python
# 실수형 -> 정수형
print(int(3.8))
````

````python
# 정수형 -> 실수형
print(float(3))
````

  - `str()` : 문자형으로 변환

````python
print(int("2") + int("5"))
````

````python
print(float("1.1") + float("2.5"))
````

````python
print(str(2) + str(5))
````

### (2) return vs. print

````python
def add_print(a, b):
    print(a + b)  # 출력
    return None  # 생략
````

````python
def add_return(a, b):
    return a + b  # 반환
````

````python
# 출력 함수 실행
x = add_print(2, 3)   # 출력: 5, 반환: x는 None, x = None
````

````python
print(x)  # x: None
````

````python
y = add_return(2, 3)  # 출력: 없음, 반환: y는 5, y = 5
````

````python
print(y)  # y: 5
````

### (4) break문(반복 제어문)
- 만약 while문의 조건 부분과 상관없이 반복문에서 나오고 싶을 때

`if`와 `break`를 함께 사용하면 **조건이 참일 때 반복문을 즉시 종료**할 수 있습니다.

- `if`: 조건문, 조건이 참(`True`)인지 검사
- `break`: 현재 실행 중인 반복문을 바로 빠져나온다.

````python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
````

````python
i = 100

while True:
    if i % 23 == 0:
        break
    i = i + 1

print(i)
````

### (5) continue문(반복 제어문)

**continue**
- 현재 반복을 건너뛰고 다음 반복으로 넘어가라는 의미

**break vs. continue 예**
- break는 게임을 아예 종료하기
- continue는 이번 라운드만 스킵하고 다음 라운드 진행하기

````python
i = 0

while i < 15:
    i = i + 1
    if i % 2 == 1:
        continue
    print(i)
````
