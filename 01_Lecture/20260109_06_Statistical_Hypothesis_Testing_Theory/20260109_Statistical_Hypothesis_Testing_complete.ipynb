# 오차와 신뢰구간

## 오차

- 통계학적 의미의 오차(error)
- 표본을 통해 얻은 평균 또는 비율이 참값(모집단의 실제 값)에서부터 어떤 확률과 정도로 벗어나 있는지에 대한 지표
- 표본을 통해 모집단을 파악하고 싶지만, 표본의 평균이 모집단의 평균과 완벽히 일치하는 경우는 거의 없음
- 때문에, 표본을 통해 모집단이 존재할 만한 범위를 추정
- 이때 오차라는 개념이 필요

- 오차는 수집된 데이터의 수와 해당 데이터의 불규칙성에 따라 달라짐
- 평균의 표준오차(Standard Error)

$$
SE = \frac{s}{\sqrt{n}}
$$

- $s$는 수집된 데이터의 표준편차 => 데이터의 불규칙성
        -표준편차가 크다는 것은 데이터의 불규칙성도 크다는 것.
- $n$은 표본 크기 => 데이터의 수

- 데이터의 수가 많을 수록 = `표본의 크기가 커질수록`
- 수집된 데이터의 불규칙성이 적을 수록 = `표준편차가 작을수록`
- 표본평균의 변동폭이 달라진다 = `표준오차가 작아진다`

- 실무에서는 반복적으로 표본을 뽑아 조사하기는 힘듭니다
- 표본을 한 번 뽑아 평균을 계산
- 그 표본평균은 분포 내 어딘가에 위치하게 될 것이라고 가정
- 표준오차가 작을수록 모집단의 참값에 더 가까워질 수 있다

## 신뢰구간

- 신뢰구간으로 참값의 범위 추정하기
- 표본 데이터를 통해서, 모집단의 실제값은 ~할 것이다 라고 주장
- 이 때, 신뢰구간이라는 개념이 필요

- 신뢰구간(CI, Confidence Interval)
: 모집단의 값이 어느 범위에 있는지 확률적으로 표현하는 방법
- 95% 신뢰구간이라는 표현을 자주 사용
- 95% 확률로 해당 구간에 모집단의 값이 포함되어 있다는 의미를 가짐
- 표본평균이 정규분포를 따를 때,
- 표본평균 +- 1.96SE 공식으로 계산
- 표본을 통해 얻은 표본평균값에서 1.96SE 범위 내에 모평균이 들어갈 확률이 95% 라는 의미

````python
np.random.seed(30)
sample_size = 30
pop_mean = 50
pop_std = 10

# 모집단에서 랜덤 표본 생성
sample = np.random.normal(pop_mean, pop_std, sample_size)
````

# t-검정(t-test)

## [1] 일표본 t-검정(One Sample t-test)

- 가정 : 모집단이 정규분포를 따른다. / 모집단의 표준편차를 모른다. 표본 표준편차로 추청
- 검정 : 단일 모집단에서 관심이 있는 연속형 변수의 평균을 특정 값과 비교한다.
  - 귀무가설 : 모평균이 특정 기준값과 같다.
  - 대립가설 : 모평균이 특정 기준값과 같지 않다.

### 예 : 정규성을 만족하는 경우

예) iris 데이터에서 setosa 품종의 sepal_length 평균이 5cm인지 아닌지를 통계적으로 검정하고, 결과를 해석하세요.(양측 검정, 유의수준=0.05)

- 귀무가설 : mu = 5
- 대립가설 : mu != 5

````python
# 데이터 불러오기
import seaborn as sns

iris_df = sns.load_dataset('iris')
iris_df.head()
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다


사용검정 : 샤피로-윌크 검정(Shapiro-Wilk test)

- 검정통계량
$$
W = \frac{\left( \sum_{i=1}^{n} a_i \, x_{(i)} \right)^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

$x_{(i)}$​,오름차순으로 정렬된 표본 값  
$x_i$​,정렬 전 표본 값들    
$\bar{x}$,표본 평균   
$a_i$​,샤피로-윌크 계수 n에 따라 미리 계산된 상수"     
n,표본 크기

- p-value
    - n ≤ 50 정도 → 정확한 표(Shapiro-Wilk table) 이용 or 근사식
    - n이 클 때 → Royston의 근사 알고리즘 등 사용 (scipy, R 등 대부분 이 방식)


````python
# iris 데이터의 setosa 품종의 sepal_length의 정규성 검정 : Shapiro-Wilk Test
import scipy.stats as stats

stats.shapiro(iris_df[iris_df['species'] == 'setosa']['sepal_length'])
````

#### 단일 표본 검정

- 귀무가설 : mu = 5
- 대립가설 : mu != 5

````python
# iris 데이터의 setosa 품종의 sepal_length의 단일 표본 검정
mu = 5

stats.ttest_1samp(iris_df[iris_df['species'] == 'setosa']['sepal_length'], mu)
````

- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다.
- iris 데이터에서 setosa 품종의 sepal_length의 평균이 5cm라고 할만한 충분한 근거가 있다.

### 예 : 정규성을 만족하지 않는 경우

예) tips 데이터에서 손님들이 주는 팁(tip)의 평균이 3달러인지 아닌지를 통계적으로 검정하고, 결과를 해석하세요.(양측 검정, 유의수준=0.05)

| 컬럼명      | 설명         |
|-------------|--------------|
| total_bill  | 총 식사 금액 |
| tip         | 팁 금액      |

| time        | 점심/저녁    |
| size        | 일행 인원 수 |

- H0 : mu = 3
- H1 : mu != 3

````python
# 데이터 불러오기
import seaborn as sns

tips_df = sns.load_dataset('tips')
tips_df.head()
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
# tips 데이터의 tip의 정규성 검정 : Shapiro-Wilk Test
import scipy.stats as stats

stats.shapiro(tips_df['tip'])
````

- p-value < 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 있다. 따라서, 정규성을 만족하지 않으므로 윌콕슨 부호 순위 검정을 진행한다.

#### 윌콕슨 부호 순위 검정

- H0 : mu = 3
- H1 : mu != 3

````python
# tips 데이터의 tip의 윌콕슨 부호 순위 검정
mu = 3

stats.wilcoxon(tips_df['tip'] - mu)
````

 - p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다.
 - 손님들이 주는 팁(tip)의 평균이 3달러고 할만한 충분한 근거가 있다.

## [2] 대응 표본 t-검정(Paired Sample t-test)

- 가정 : 모집단이 정규분포를 따른다.
- 검정 : 단일 모집단에 대해 처리를 가했을 때, 두 조건(전 vs 후, 처치 A vs B 등)의 평균 차이를 검정한다.
  - 귀무가설 : 두 모평균 차이가 0이다.
  - 대립가설 : 두 모평균 차이가 0이 아니다.

### 예 : 정규성을 만족하는 경우

예) 10명의 참가자에게서 학습 효과를 확인하기 위해 학습 프로그램 참여 전과 후의 점수를 비교하였다. 각 참가자의 학습 프로그램 참여 전과 후 평균 점수에 차이가 있는지 검정하세요.(양측 검정, 유의수준=0.05)

````python
# 데이터 불러오기
import pandas as pd

exam_dict = {'before' : [60, 62, 65, 63, 67, 70, 72, 68, 66, 64],
        'after' : [63, 65, 69, 66, 71, 73, 77, 72, 68, 69]}

exam_df = pd.DataFrame(exam_dict)
exam_df
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
exam_df['diff'] = exam_df['after'] - exam_df['before']
````

````python
exam_df.head()
````

````python
# exam_df의 diff의 정규성 검정 : Shapiro-Wilk Test
import scipy.stats as stats

stats.shapiro(exam_df['diff'])
````

- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다. 따라서, 정규성을 만족한다.

#### 대응 표본 검정

- 귀무가설 : D = 0
- 대립가설 : D != 0

````python
# 대응 표본 검정
stats.ttest_rel(exam_df['before'], exam_df['after'])
````

- p-value < 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 있다.
- 각 참가자의 학습 프로그램 참여 전과 후 평균 점수에 차이가 있다.
- after의 점수가 before 점수보다 높으므로 학습 프로그램이 효과가 있다.

````python
# 학습 프로그램 전후 시험 점수 평균 계산
exam_df.mean()
````

### 예 : 정규성과 등분산성을 만족하는 경우

예) 한 연구자는 두 가지 학습 방법(A, B)이 학생들의 시험 점수에 차이를 주는지 알아보고자 한다. 학습 방법(A, B)에 따른 평균 시험 점수에 차이가 있는지 검정하세요.(양측 검정, 유의수준=0.05)

- 귀무가설 : 학습 방법(A, B)에 따른 그룹A와 그룹B의 평균 시험 점수에 차이가 없다.
- 대립가설 : 학습 방법(A, B)에 따른 그룹A와 그룹B의 평균 시험 점수에 차이가 있다.

````python
# 데이터 준비하기
import pandas as pd
import numpy as np

exam_dict = {'group_a' : [54, 52, 46, 61, 42, 43, 54, 52, 46, 51, 45, 53, 45, 55, 53, 55, 57, 47, 49, 56, 49, 52, 50, 56, 49, 47, 60, 54, 55, 55],
           'group_b' : [48, 51, 51, 56, 58, 51, 57, 54, 49, 52, 51, 49, 64, 65, 58, 60, 55, 61, 49, 51, 63, 47, 56, 45, 57, 46, 59, 62, 60, 56]}

exam_df = pd.DataFrame(exam_dict)
exam_df.head()
````

#### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.

````python
# 정규성 검정
import scipy.stats as stats

print(stats.shapiro(exam_df['group_a']))
print(stats.shapiro(exam_df['group_b']))
````

- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다.
- 따라서, 정규성을 만족한다.

#### 등분산성 검정

- 귀무가설 : 등분산을 따른다.
- 대립가설 : 등분산을 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 등분산을 따른다고 할만한 충분한 근거가 있다)

````python
# 등분산성 검정
import scipy.stats as stats

stats.levene(exam_df['group_a'], exam_df['group_b'])
````

- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다.
- 따라서, 등분산성 만족한다.

#### 독립 표본 검정

- 귀무가설 : 학습 방법(A, B)에 따른 그룹A와 그룹B의 평균 시험 점수에 차이가 없다.
- 대립가설 : 학습 방법(A, B)에 따른 그룹A와 그룹B의 평균 시험 점수에 차이가 있다.

````python
stats.ttest_ind(exam_df['group_a'], exam_df['group_b'], equal_var=True)
````

- p-value < 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 있다.
- 따라서, 학습 방법(A, B)에 따른 그룹A와 그룹B의 평균 시험 점수에 차이가 있다.

````python
exam_df.mean()
````

# 카이제곱검정(Chi-Square Test)

| 구분     | 설명                                                                 |
|----------|----------------------------------------------------------------------|
| 적합도 검정 | 한 변수의 관측 분포가 기대 분포와 일치하는지 검정                     |
| 독립성 검정 | 두 변수 간 관계가 독립적인지 검정               |
| 동질성 검정 | 여러 집단 간 분포가 동일한지 비교     |

**카이제곱 검정(Chi-Square Test)** : 교차표를 기반으로 관찰빈도와 기대빈도의 차이가 통계적으로 유의한지 검정하는 방법
  - 관찰빈도: 실제 자료로부터 얻은 각 셀의 빈도
  - 기대빈도: 두 변수가 독립적이라고 가정했을 때 이론적으로 기대되는 빈도

**카이제곱 검정의 세 가지 종류**

1. 적합도 검정(Goodness-of-fit Test)
- 설명: 하나의 범주형 변수의 관측 분포가 이론적 분포(기대 분포)와 일치하는지 검정

# [1] 적합도 검정

적합도 검정(Goodness-of-fit Test)
- 설명: 하나의 범주형 변수의 관측 분포가 이론적 분포(기대 분포)와 일치하는지 검정
- 예시: 주사위를 굴렸을 때, 각 주사위의 값이 1/6의 확률로 등장하는 게 맞는지 검정

```Python
scipy.stats.chisquare(observed, expected)
```

예) titanic 데이터에서 sex 변수에 대한 분할표를 생성하고 아래 가설에 대하여 적합도 검정을 수행하세요.

- 귀무가설 : titanic 데이터에서 남성의 비율은 50%이고, 여성의 비율은 50%이다.
- 대립가설 : titanic 데이터에서 남성의 비율은 50%이고, 여성의 비율은 50%라고 할 수 없다.

````python
# 데이터 불러오기
import seaborn as sns

titanic_df = sns.load_dataset('titanic')
titanic_df.head()
````

````python
# observed
table = titanic_df['sex'].value_counts()
table
````

````python
# expected
len(titanic_df['sex']) / 2
````

````python
# 적합도 검정
from scipy.stats import chisquare

chisquare(table, len(titanic_df['sex']) / 2)
````

- p-value < 유의수준 : 유의수준 하에서 귀무가설 기각할만한 충분한 근거가 있다.
- 따라서, titanic 데이터에서 남성의 비율은 50%이고, 여성의 비율은 50%라고 할 수 없다.

# [2] 독립성 검정

독립성 검정(Test of Independence)
- 설명: 두 범주형 변수 간에 통계적으로 독립적인지(연관성이 없는지) 검정
- 예시: 성별(남/여)과 흡연 여부(흡연/비흡연)가 서로 독립적인지 확인

```Python
pd.crosstab()
scipy.stats.chi2_contingency(table)
correction=True : 2x2 표에서 보수적 검정 (기본, 소표본)
correction=False : 보정 없이 검정 (대표본인 경우 권장)
```

예) titanic 데이터에서 class(좌석등급)과 survived(생존 여부)에 대한 분할표를 생성하고 서로 독립인지 확인하기 위하여 독립성 검정을 수행하세요.

- 귀무가설 : class(좌석등급)과 survived(생존 여부)는 독립이다.
- 대립가설 : class(좌석등급)과 survived(생존 여부)는 독립이 아니다.

````python
titanic_df.head()
````

````python
# 분할표 생성하기
import pandas as pd

table = pd.crosstab(titanic_df['class'], titanic_df['survived'])
table
````

````python
# 독립성 검정
from scipy.stats import chi2_contingency

chi, p, dof, exp = chi2_contingency(table)

print('Statistic: ', chi)
print('p-value: ', p)
print('Degree of Freedom: ', dof)
print('Expected: \n', exp)


````

- 귀무가설 : class(좌석등급)과 survived(생존 여부)는 독립이다.
- 대립가설 : class(좌석등급)과 survived(생존 여부)는 독립이 아니다.

- p-value < 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 있다.
- 따라서, class(좌석등급)과 survived(생존 여부)는 독립이 아니다.

# [0] 분산분석(ANOVA)

- 가정 : 독립성, 정규성, 등분산성
- 검정 : 세 개 이상의 집단에서 그룹 평균 간 차이를 그룹 내 변동에 비교하는 통계 분석 방법
  - 귀무가설 : k개 집단의 모평균은 모두 같다.
  - 대립가설 : k개 집단의 모평균이 모두 같다고 할 수는 없다.

- 일원 배치 분산분석(One-way ANOVA) : 독립변수 1개, 종속변수 1개
- 이원 배치 분산분석(Two-way ANOVA) : 독립변수 2개, 종속변수 1개
- 다원 배치 분산분석(Multi-way ANOVA) : 독립변수 3개 이상, 종속변수 1개

- 사후 검정(Post-hoc Test) : 분산분석 결과 귀무가설이 기각되어 적어도 한 집단에서 평균의 차이가 있음이 통계적으로 증명된 경우, 어떤 집단들에 대해서 평균의 차이가 존재하는지 알아보기 위해 실시하는 분석

# [1] 일원 배치 분산분석

## 예 : 정규성과 등분산성을 만족하는 경우

예) iris 데이터에서 세 가지 붓꽃 품종별(setosa, versicolor, virginica)로 평균 sepal_width에 차이가 있는지 검정하세요.(유의수준=0.05)

- 귀무가설 : 세 가지 종의 sepal_width의 평균은 모두 같다.
- 대립가설 : 세 가지 종의 sepal_width의 평균이 모두 같은 것은 아니다.

````python
# 데이터 불러오기
import seaborn as sns

iris_df = sns.load_dataset('iris')
iris_df.head()
````

````python
# 범주의 종류의 개수
iris_df['species'].nunique()
````

````python
# 범주의 종류
iris_df['species'].unique()
````

### 정규성 검정

- 귀무가설 : 데이터가 정규분포를 따른다.
- 대립가설 : 데이터가 정규분포를 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
# 세 가지 품종별 sepal_width
setosa = iris_df[iris_df['species'] == 'setosa']['sepal_width']
versicolor = iris_df[iris_df['species'] == 'versicolor']['sepal_width']
virginica = iris_df[iris_df['species'] == 'virginica']['sepal_width']
````

````python
# 정규성 검정
import scipy.stats as stats

print(stats.shapiro(setosa))
print(stats.shapiro(versicolor))
print(stats.shapiro(virginica))
````

- p-value > 유의수준 : 귀무가설 기각X, 정규성 만족

### 등분산성 검정

- 귀무가설 : 등분산을 따른다.
- 대립가설 : 등분산을 따르지 않는다.
- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 등분산을 따른다고 할만한 충분한 근거가 있다)

````python
# 등분산성 검정
import scipy.stats as stats

stats.levene(setosa, versicolor, virginica)
````

- p-value > 유의수준 : 귀무가설 기각X, 등분산성 만족

### 일원 배치 분산분석

- 귀무가설 : 세 가지 종의 sepal_width의 평균은 모두 같다.
- 대립가설 : 세 가지 종의 sepal_width의 평균이 모두 같은 것은 아니다.

````python
import scipy.stats as stats

stats.f_oneway(setosa, versicolor, virginica)
````

- p-value < 유의수준 : 귀무가설 기각, 세 가지 종의 sepal_width의 평균이 모두 같은 것은 아니다.

### 사후 검정

- 모든 집단 수준에 대하여 두 집단씩 짝을 지어 각각 다중 비교를 수행한다.
- 귀무가설 : 집단 간 평균은 같다.
- 대립가설 : 집단 간 평균은 같지 않다.

````python
# 사후 검정
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(data=iris_df['sepal_width'], groups=iris_df['species'])
tukeyhsd = mc.tukeyhsd(alpha=0.05)
tukeyhsd.summary()
````

````python
fig = tukeyhsd.plot_simultaneous()
````

````python
setosa.mean(), versicolor.mean(), virginica.mean()
````

- 귀무가설 : 세 집단(A, B, C)의 평균 점수는 모두 같다.
- 대립가설 : 세 집단(A, B, C)의 평균 점수가 모두 같은 것은 아니다.

````python
# 데이터 준비하기
import pandas as pd
import numpy as np

program_a = [49, 45, 49, 50, 50, 52, 45, 47, 54, 47, 52, 48, 49, 50, 47, 48, 50, 45, 48, 52, 49, 44, 48, 49, 46, 51, 47, 47, 50, 49]
program_b = [44, 58, 52, 51, 53, 64, 41, 63, 46, 52, 46, 55, 54, 50, 47, 49, 39, 58, 53, 50, 55, 45, 41, 54, 53, 60, 44, 34, 38, 49]
program_c = [61, 74, 49, 53, 58, 36, 61, 52, 56, 56, 70, 63, 58, 55, 46, 59, 44, 63, 38, 82, 67, 39, 63, 43, 49, 50, 36, 41, 49, 72]

program_df = pd.DataFrame({
    'value' : np.concatenate([program_a, program_b, program_c]),
    'program' : ['A']*30 + ['B']*30 + ['C']*30
    })

program_df.head()
````

````python
# 정규성 검정
import scipy.stats as stats

print(stats.shapiro(program_a))
print(stats.shapiro(program_b))
print(stats.shapiro(program_c))
````

- p-value > 유의수준 : 귀무가설 기각X, 정규성 만족

### 등분산성 검정

- 귀무가설 : 등분산을 따른다.
- 대립가설 : 등분산을 따르지 않는다.

- p-value > 유의수준 : 유의수준 하에서 귀무가설을 기각할만한 충분한 근거가 없다(유의수준 하에서 데이터가 정규분포를 따른다고 할만한 충분한 근거가 있다)

````python
# 정규성 검정
import scipy.stats as stats

print(stats.shapiro(group_a))
print(stats.shapiro(group_b))
print(stats.shapiro(group_c))
````

- 운동 프로그램 A : p-value < 유의수준 : 귀무가설 기각, 정규성 만족X

- 운동 프로그램 C : p-value > 유의수준 : 귀무가설 기각X, 정규성 만족

- 대립가설 : 세 가지 운동 프로그램의 평균 체중 감소량이 모두 같은 것은 아니다.

# 로지스틱 회귀**로지스틱 회귀분석**

- 로지스틱 회귀분석 : 종속변수(y)가 **이진(binary)** 또는 **범주형**(주로 0/1)일 때, 독립변수(X)들이 y가 1이 될 **확률**을 모델링하는 기법  
  → 선형회귀처럼 y를 직접 예측하는 대신, **확률**을 예측하고 이를 0~1 사이로 제한

  - **로지스틱 회귀** (이진 로지스틱 회귀) : y가 0 또는 1인 경우 (가장 흔함)
  - **다항 로지스틱 회귀** (Multinomial) : y가 3개 이상의 범주인 경우
  - **서열 로지스틱 회귀** (Ordinal) : y의 범주에 순서가 있는 경우

로지스틱 회귀는 **선형 결합**을 직접 y에 넣지 않고, 로짓(logit)이라는 변환을 거침

$$
\text{logit}(p) = \ln\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1X_1 + \beta_2X_2 + \cdots + \beta_kX_k
$$

- p = P(y=1 | X) : 성공(1) 확률
- $\frac{p}{1-p}$ = **오즈(odds)** : 성공할 확률 / 실패할 확률      
- 오즈비 = exp(β) = e^β : X가 1단위 증가할 때 오즈가 β만큼 변화



````python
import pandas as pd
import statsmodels.api as sm
import numpy as np

data = sm.datasets.spector.load_pandas().data

# - GRADE: 성적이 향상되었는지 여부 (1: 향상, 0: 미향상) - 종속 변수 (이진)
# - GPA: 기존 학점 평균 
# - TUCE: 경제학 시험 점수 
# - PSI: 새로운 학습 프로그램 참여 여부 (1: 참여, 0: 미참여) 

y = data['GRADE']  
X = data[['GPA', 'TUCE', 'PSI']]  
X = sm.add_constant(X)

# statsmodels의 Logit 사용
model = sm.Logit(y, X)
result = model.fit()
print(result.summary())

# 오즈비 = exp(계수) → 해당 변수가 1단위 증가할 때 오즈가 몇 배 변하는지
odds_ratios = np.exp(result.params)
print(odds_ratios) # TUCE 오즈비가 1에 가까움 = 성공할확률 == 실패할확률 => 큰의미가 없는 변수
````

````python
new_data = pd.DataFrame({'const': [1], 'GPA': [3.5], 'TUCE': [20], 'PSI': [1]})
predicted_prob = result.predict(new_data)
print(f"예측 확률: {predicted_prob[0]}") # 0.5 기준으로 label 변경
````

````python
# IQR 기반 이상치 처리하기
def remove_outliers(data, col):
    q3 = data[col].quantile(0.75)
    q1 = data[col].quantile(0.25)

    iqr = q3 - q1

    lower_limit = q1 - 1.5*iqr
    upper_limit = q3 + 1.5*iqr

    return data[(data[col] > lower_limit) & (data[col] < upper_limit)]
````

````python
# Fare 컬럼 이상치 처리하기
print("이상치 처리 전: ", data.shape)
data = remove_outliers(data, 'Fare')
print("이상치 처리 후: ", data.shape)
````
