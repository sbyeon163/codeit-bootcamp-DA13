## 불린 인덱싱(마스킹)

- 정의: 조건식의 결과로 나온 `True/False`값을 이용해, 원하는 데이터만 선택하는 방법
- 원리: 조건을 만족하는 행(또는 열)은 `True`로 표시하고, 조건을 만족하지 않으면 `False`로 숨김
- 예시: 특정 값 이상인 데이터, 특정 조건을 만족하는 행만 필터링

(예시1)

gdp_array = np.array([6610, 7637, 8127, 8885, 10385, 12565, 14403, 12398, 8282, 10672])
gdp_array

gdp 10000보다 큰 경우만 출력
```python
gdp_array > 10000 # mask
gdp_array[gdp_array > 10000]
```
(예시2)

두가지 조건
10000보다 크거나 8000보다 작은 경우
```python
mask1 = gdp_array >10000
mask2 = gdp_array <8000

gdp_array[mask1 | mask2] # or == |       and == &
gdp_array[(gdp_array > 10000) | (gdp_array < 8000)] # 괄호로 각 마스크를 닫아줘야함
```
(예시3)

8000이상이고 10000 이하인 경우
```python
mask1 = gdp_array >= 8000
mask2 = gdp_array <= 10000

gdp_array[mask1 & mask2]
```

## (7) 그룹화

- `groupby`: 데이터 프레임에서 특정 컬럼을 기준으로 그룹을 나눈 뒤, 각 그룹별로 집계(합계, 평균 등), 변환, 필터링 등을 수행

````python
df = pd.read_csv("../data/tottenham_2021.csv", index_col=0)
# 10골 이상 선수 legends 컬럼에 yes / 미만인 선수들은 no
# NaN -> None
df.loc[df['assists'] >=10 ,'legends'] = 'yes'
df.loc[df['assists'] < 10 ,'legends'] = 'no'

# legends yes / no 그룹의 goals 평균
# dataframe.groupby([그룹할 컬럼])[타겟컬럼].연산메소드

df.groupby(['legends'])['goals'].mean()
````
