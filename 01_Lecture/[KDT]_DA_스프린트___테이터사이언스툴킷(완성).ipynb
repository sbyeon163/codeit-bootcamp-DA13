## array

- `np.zeros(n)` : 0이 `n`개 들어있는 배열 생성  
- `np.arange(a)` : 0 이상 `a` 미만의 정수 배열 생성  
- `np.arange(a, b)` : `a` 이상 `b` 미만의 정수 배열 생성  
- `np.arange(a, b, c)` : `a` 이상 `b` 미만, `c` 간격의 배열 생성

```
# 라이브러리 불러오기
import numpy as np
```

```
array1 = np.array([1, 3, 5, 6, 7, 9])
array1
```

```
type([1, 3, 5, 6, 7, 9])
```

```
type(array1)  # ndarray(n dimensional array) : n차원 배열
```

```
# np.zeros(n) : 0이 n개 들어있는 배열 생성
np.zeros(5)
```

```
# np.arange(a) : 0 이상 a 미만의 정수 배열 생성
np.arange(10)
```

```
# np.arange(a, b) : a 이상 b 미만의 정수 배열 생성
np.arange(2, 10)
```

```
# np.arange(a, b, c) : a 이상 b 미만, c 간격의 배열 생성
np.arange(4, 17, 3)
```

## (4) 데이터 일부 선택하기

**데이터 선택 방법**

1. `iloc` (integer location): 행/열 **번호**로 접근하여 정수 기반 인덱스를 사용해 데이터 선택
2. `loc` (location): 행/열 **이름**으로 접근하여 레이블(이름) 기반 인덱스를 사용해 데이터 선택
3. 대괄호(`[]`) 사용
    - 열 선택: `df['column1']` → `'column1'` 열 반환  
    - 여러 열 선택: `df[['column1', 'column2']]` → 두 개 이상의 열 반환  
    - 행 슬라이싱: `df[0:3]` → 0~2번째 행 반환 (정수 위치 기준)
